```python
comp = []
avg = []
for i in range( 0, 100 ):
    temp = []
    fArray = data[ np.random.choice( [0, 1, 2, 3] ) ]
    fArray = fArray + ( np.random.choice( record[3] ) - np.average( fArray ) )
    for a in fArray:
        if a > 5.25:
            temp.append( a )
    fArray = np.array( temp )
    avg.append( np.average( fArray ) )
    comp.append( fArray )
```

```python
for x in comp:
    plt.hist( x, bins=22, alpha=0.3 )
```

##Average Tournament

```python
avgTrials = []
for a in range( 0, 3 ):
    winners = []
    wins = 0
    loss = 0
    for i in range( 0, 10000 ):
        trialWin = []
        for x in comp:
            temp = np.random.choice( x, size=5 )
            temp = np.delete( temp, [temp.tolist().index( np.amin( temp ) ), temp.tolist().index( np.amax( temp ) ) ] )
            trialWin.append( np.average( temp ) )
        winners.append( trialWin.index( np.amin( trialWin ) ) )
        if avg.index( np.amin( avg ) ) == trialWin.index( np.amin( trialWin ) ):
            wins += 1
        else:
            loss += 1
    avgTrials.append( [wins, loss] )
```

##Best Of Tournament

```python
bestTrials = []
for a in range( 0, 3 ):
    winners = []
    bWins = 0
    bLoss = 0
    for i in range( 0, 10000 ):
        trialWin = []
        for x in comp:
            temp = np.random.choice( x, size=5 )
            #temp = np.delete( temp, [temp.tolist().index( np.amin( temp ) ), temp.tolist().index( np.amax( temp ) ) ] )
            trialWin.append( np.amin( temp ) )
        winners.append( trialWin.index( np.amin( trialWin ) ) )
        if avg.index( np.amin( avg ) ) == trialWin.index( np.amin( trialWin ) ):
            bWins += 1
        else:
            bLoss += 1
    bestTrials.append( [bWins, bLoss] )
```
