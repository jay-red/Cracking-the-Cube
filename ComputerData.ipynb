```python
comp = []
avg = []
for i in range( 0, 100 ):
    l = []
    for i in range( 0, 1000 ):
        l.append( np.random.choice( tData ) )
    na = np.array( l ) + np.random.random() * 10 - 5
    comp.append( na )
    avg.append( np.average( na ) )
```

```python
wins = 0
loss = 0
for i in range( 0, 100 ):
    winners = []
    for t in range( 0, 1000 ):
        trialWin = []
        for x in comp:
            trialWin.append( np.average( np.random.choice( x, size=5 ) ) )
        winners.append( trialWin.index( np.amin( trialWin ) ) )
    if np.amin( avg ) == avg[ np.argmax( np.bincount( winners ) ) ]:
        wins += 1
    else:
        loss += 1
   
```
